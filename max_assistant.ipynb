```python
import speech_recognition as sr
import pyaudio
from gtts import gTTS
import random, os
import pyttsx3
import webbrowser

engine = pyttsx3.init()

def listen():
    
    mic_name = "Microphone (Realtek HD Audio Mic input)"
    sample_rate = 8000
    chunk_size = 2048
    r = sr.Recognizer()


    mic_list = sr.Microphone.list_microphone_names()
    for i, microphone_name in enumerate(mic_list):
        if microphone_name == mic_name:
            device_id = i

    with sr.Microphone(sample_rate = sample_rate, chunk_size = chunk_size) as source:
        #r.adjust_for_ambient_noise(source)
        #print("Say something")
        #speak("Say something")
        audio = r.listen(source)

        try:
            text = r.recognize_google(audio)
            print(text)
            #speak("you said " + text)
            return text
            
        except sr.UnknownValueError:
            print("couldnt understand")

        except sr.RequestError as e:
            print("Error")

def main():
    
        a = r.listen(source)
        #print('3')
        
        try:
            text = r.recognize_google(a)
            #print('4')
            print(text)
            keyword = text.split()
            
            if text == "max":
                online()
                
            elif text == "introduce yourself":
                online1()
                
            elif text in ['music','play some music','play some melody','play a song','i want some music','song']:
                media()
                
            elif text in ['hey max','wake up max','wake up','oyee']:
                getonline()
                
            elif text == "down":
                gooffline()
                
            elif text in ['shutdown','see you tomorrow','bye max','bye','ok bye','see you tomorrow max','ok max see you tomorrow','see you after some time','see you later','have some sleep max']:
                shutdown()
                
            elif text in ['restart the system']:
                restart()
         
            elif text in ['hi','i','hey','hello']:
                d = random.choice(['hey','hi','hello'])
                speak(d)
            else:
                for i in keyword:
                    if i in ['music', 'media', 'entertainment', 'song']:
                        media()
                        break 
                    elif i in ['hi','i','hey','hello','hay']:
                        d = random.choice(['hey','hi','hello'])
                        speak(d)
                        break
                    elif i in ['introduce','who']:
                        online1()
                        break
                    elif i in ['search']:
                        search(text)
                        break
                    elif i in ['what','do','functions','capable','tasks','help']:
                        tasks()
                        break
#                     else:
#                         search(text)
#                         break
            
        except sr.UnknownValueError:
            print("couldnt understand")

        except sr.RequestError as e:
            print("Error")
            
        
        #speak('hi.......what can i do for you? .......!!')

def online():
    g = random.choice(['hey','hi','hello'])
    l = random.choice(['How may I help you','what can I do for you','please tell me'])
    speak(g+'...'+l)
#     tts=_tts()
#     tts.start(g+'...'+l)
#     del(tts)

def search(text):
    key=text.split()
    k=key.index('search')
    srch=''
    for i in range(k+1,len(key)):
        srch=srch+' '+key[i]
    
        
    url = "https://www.google.com/search?q={}".format(srch)
    speak("Using google search")
    webbrowser.open_new_tab(url)
    
def online1():
    speak('hi...i am max....... your personnel assistant....')
#     tts=_tts()
#     tts.start('hi...i am max....... your personnel assistant....')
#     del(tts)   
    
def media():
    f=random.choice(['that will be good for this time!!!!','thats a good choice!!!!','good choice!!!!!'])
    speak(f)                               
    speak('connecting to player')                 
    os.system('starting kmplayer.exe')
    speak('which song do you want to play')
    a = r.listen(source)
    text = r.recognize_google(a)
    try:
        os.startfile('F:\Music\Eng. Songs\\'+text+'.mp3')
        speak('playing '+text+' for you')
    except os.error:
        speak('Sorry.... do not have this song')
    
def shutdown():
    speak('ok')
    speak('connecting to command prompt')
    speak('shuttting down')
    os.system('shutdown -s')

def restart():
    speak('ok')
    speak('closing all applications')
    os.system('restart -r')

def gooffline():
        speak('ok')
        speak('closing all systems')
        quit()

#def getonline():
#    speak('ok')
#    os.system('start C:/Rainmeter.exe') 
#    speak('starting systems')
#    speak('i am online')

def tasks():
#     tts=_tts()
#     tts.start('I can do the following tasks for you... I can play music of your choice.... I can restart the system.... I can also shut it down for you.... other functions are in the developing stage and will be available soon')
#     del(tts)  
      speak('I can do the following tasks for you... I can play music of your choice...Can search your query online... Can restart the system... I can also shut it down for you... other functions are in the developing stage and will be available soon')

def speak(text):
#     myob = gTTS(text=text,lang='en',slow=False)
#     myob.save("test.mp3")
#     os.system("mpg321 test.mp3")
    engine.say(text)
    print(text)
    engine.runAndWait()
    #r.adjust_for_ambient_noise(source)

# class _tts: 
#     engine = None
#     rate = None
#     def __init__(self):
#         self.engine = pyttsx3.init()
#     def start(self,text_):
#         self.engine.say(text_)
#         self.engine.runAndWait()
    
if __name__ == "__main__":
    print('say')
    r = sr.Recognizer()
    print('something')
    with sr.Microphone() as source:
        while 1:
            main()
            #print('2')

```
